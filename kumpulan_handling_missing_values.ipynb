- Mengimport Libraries Yang Dibutuhkan

```python
import seaborn as sns
import pandas as pd
```

- Menyimpan Data Ke Variabel Tertentu Dan Menampilkannya

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

- Menampilkan Info Data

```python
titanic.info()
```

- Menampilkan Berapa Banyak Data Yang Kosong

```python
titanic.isnull().sum()
```

- Mengubah Tipe Data Yang Ada Ke Tipe Data Angka Dan Setelah Diubah, Data Kosong Diubah Dengan Metode 'IterativeImputer', Dan Menampilkan Berapa Banyak Data Yang Kosong

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import IterativeImputer

iterative_imputer = IterativeImputer()
orr = OrdinalEncoder()

# orr_column = ["sex", "embarked", "class", "who", "adult_male", "deck", "embark_town", "alive", "alone"]
orr_column = titanic.select_dtypes(["object", "category", "bool"]).columns
titanic_null = titanic.isnull().sum()
# iterative_imputer_column = ["age", "embarked", "deck", "embark_town"]
iterative_imputer_column = list(titanic_null.index)

titanic[orr_column] = orr.fit_transform(titanic[orr_column])
titanic[iterative_imputer_column] = iterative_imputer.fit_transform(titanic[iterative_imputer_column])

# titanic = pd.DataFrame(titanic, columns=titanic.columns)

titanic.isnull().sum()
```

- Mengubah Tipe Data 'float64' Ke 'int64'

```python
# astype_column = ["sex", "age", "embarked", "class", "who", "adult_male", "deck", "embark_town", "alive", "alone"]
astype_column = titanic.select_dtypes("float64").columns

titanic[astype_column] = titanic[astype_column].astype("int64")
titanic.head()
```
