- Mengimport Libraries Yang Dibutuhkan

```python
import seaborn as sns
import pandas as pd
```

- Menyimpan Data Ke Variabel Tertentu

```python
tips = sns.load_dataset("tips")
```

- Menampilkan Data Diawal Dengan Default 5 Row

```python
tips.head()
```

- Mengubah Data Menggunakan 'OrdinalEncoder'

```python
from sklearn.preprocessing import OrdinalEncoder

column = ["sex", "smoker", "day", "time"]
orr = OrdinalEncoder()
orr_result = orr.fit_transform(tips[column])
tips[column] = pd.DataFrame(orr_result, columns=column)
tips.head()
```

- Mengubah Data Menggunakan Metode 'MinMaxScaler'

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
min_max_scaler_result = min_max_scaler.fit_transform(tips)
pd.DataFrame(min_max_scaler_result, columns=tips.columns).head()
```

- Mengubah Data Menggunakan Metode 'StandardScaler'

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
standard_scaler_result = standard_scaler.fit_transform(tips)
pd.DataFrame(standard_scaler_result, columns=tips.columns).head()
```

- Mengubah Data Menggunakan Metode 'RobustScaler'

```python
from sklearn.preprocessing import RobustScaler

robust_scalar = RobustScaler()
robust_scalar_result = robust_scalar.fit_transform(tips)
pd.DataFrame(robust_scalar_result, columns=tips.columns).head()
```
