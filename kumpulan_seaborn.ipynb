- Mengimport libraries Tertentu Yang Dibutuhkan

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- Menyimpan Data Ke Variabel Tertentu

```python
titanic = sns.load_dataset("titanic")
fmri = sns.load_dataset("fmri")
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
```

- Melihat Data Dibedakan Warna Sesuai Data 'sex', Bentuk Sesuai Data 'class', Dan Ukuran Sesuai Data 'class' Dengan Markers, Warna, Ukuran Tertentu

```python
marker = {"First": "D", "Second": "o", "Third": "X"}
sex = {"male":"red", "female":"blue"}
plt.figure(figsize=(5, 5))
sns.scatterplot(data=titanic, x="age", y="fare", hue="sex", style="class", markers=marker, palette=sex, size="class", sizes=(25, 80), alpha=0.5)
```

- Melihat Data Dibedakan Dengan Column 'class', Warna Disesuaikan Data 'sex', Dan Bentuk Yang Disesuaikan Data 'alive'

```python
plt.figure(figsize=(5, 5))
sns.relplot(data=titanic, x="age", y="fare", hue="sex", col="class", palette=sex, style="alive", alpha=0.5)
```

- Melihat Data Yang Berbentuk 'line'

```python
sns.set_theme(style="darkgrid")
sns.relplot(data=flights, x="month", y="passengers", kind="line")
```

- Melihat Data Yang Dibedakan Warna Sesuai Data 'event' Dan Plot Yang Berbentuk 'line'

```python
sns.relplot(data=fmri, x="timepoint", y="signal", hue="event", kind="line")
```

- Melihat Data Yang Dibedakan Warna Sesuai Data 'sex' Dan Bentuk Sesuai Data 'class' Dengan Markers Tertentu Sesuai Data 'class' Dan Data X, Data Y Yang Sama

```python
sns.relplot(data=titanic, x="age", y="fare", hue="sex", style="class", kind="line", errorbar=None, markers=marker)
```

- Melihat Data Yang Dibedakan Column 'time' Dana Warna Sesuai Data 'smoker' Dengan Data X dan Data Y Yang Sama

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col="time", alpha=0.5)
```

- Melihat Data Yang Dibedakan Bentuk Sesuai Data 'alone' Dan Warna Sesuai Data 'sex' Dengan Data X Dan Data Y Yang Sama

```python
marker = {True: "o", False: "X"}
sns.relplot(data=titanic, x="age", y="fare", hue="sex", style="alone", markers=marker, alpha=0.5)
```

- Melihat Data Yang Dibedakan Warna, Sesuai Data 'year' Dengan Data X Dan Data Y Yang sama

```python
sns.relplot(data=flights, x="month", y="passengers", hue="year", kind="line")
```

- Melihat Data Yang Terpisah Dari Row 'event' Dan Column 'region' Dengan Data X Dan Data Y Yang Sama

```python
sns.relplot(data=fmri, x="timepoint", y="signal", hue="subject", row="event", col="region", kind="line", height=4)
```

- Melihat Hubungan Antara 'total_bill' Dan 'tip' Dengan Terang (sedikit) Dan Gelap (banyak)

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex")
```

- Melihat Berapa Banyak Data Yang Direkam Dengan Dibedakan Dari Data 'day'

```python
sns.histplot(data=tips, x="day", shrink=0.8)
```

- Melihat Hubungan Antara 'total_bill' Dan 'tip' Dengan Warna Dibedakan 'day', Dan Melihat Kepadatan Data (Kepadatan bukan hanya soal “berapa banyak data di sini”—tetapi “seberapa besar kemungkinan kita menemukan data di sini”)

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde", hue="day")
```

- Melihat Regressionnya

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")
```

- Melihat Semua Hubungan Column Dengan Column Yang Lain

```python
sns.pairplot(data=titanic)
```

- Melihat Data Dengan Marginal Distributions Digambarkan Garis2

```python
sns.relplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip")
```

- Melihat Kepadatan Data (Density) Dengan Dibedakan Warna Sesuai Data 'day'

```python
sns.kdeplot(data=tips, x="total_bill", hue="day", multiple="stack")
```

- Melihat Data Proportion (bagian dari total data) Memakai Persentase (0.1 menjadi 10%)

```python
sns.displot(data=tips, x="total_bill", kind="ecdf")
```
