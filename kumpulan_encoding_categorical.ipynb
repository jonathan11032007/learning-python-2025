- Membuat Data Yang Baru, Dihapuskan Data Yang Kosong, Dan Menampilkan Berapa Data Yang Kosong

```python
new_titanic = sns.load_dataset("titanic").dropna()
new_titanic.isnull().sum()
```

- Mengubah Data Dengan Metode 'LabelEncoder'

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

deck = new_titanic[["deck"]]
deck_fit_transform = label_encoder.fit_transform(deck)
new_titanic.head()
```

- Menggantikan Data 'deck' Dengan DataFrame Yang Baru Dan Menampilkan Data Diawal Dengan Default 5 Row

```python
new_titanic[["deck"]] = pd.DataFrame(deck_fit_transform, columns=deck.columns, index = deck.index)
new_titanic.head()
```

- Menampilkan Value Sebelumnya Yang Sudah Diubah
